```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("movies.csv")
df.head()
```

```python
df['review']
```

```python
def to_lower(text):
  return text.lower()

df['review'] = df['review'].apply(to_lower)
df['review']
```

```python
import re
def html_removal(text):
  tags = re.compile("<.*>")
  return tags.sub(r'',text)


df['review'] = df['review'].apply(html_removal)
df['review']
```

```python
def remove_url(text):
  url = re.compile("\b((?:https?|ftp):\/\/)?(?:www\.)?[a-zA-Z0-9\-]+(\.[a-zA-Z]{2,})(\/[^\s]*)?\b")
  return url.sub(r'',text)

df['review'] = df['review'].apply(remove_url)
df['review']
```

```python
import string
string.punctuation
```

```python
def remove_punc(text):
  # for char in text:
  #   if char in string.punctuation:
  #     text = text.replace(char,'')
  # return text
  return text.translate(str.maketrans('','',string.punctuation))

df['review'] = df['review'].apply(remove_punc)
df['review']
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
```

```python
def text_tokenization(text):
  tokens = word_tokenize(text)
  return tokens

df['review'] = df['review'].apply(text_tokenization)
df['review']
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def stopword_removal(text):
  stop_words = set(stopwords.words('english'))
  return [word for word in text if word not in stop_words]

df['review'] = df['review'].apply(stopword_removal)
df['review']
```

```python
# from nltk.stem import PorterStemmer
# ps = PorterStemmer()

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# def stemming(text):
#   return [ps.stem(word) for word in text]

# df['review'] = df['review'].apply(stemming)
# df['review']

def lemmatization(text):
  return [WordNetLemmatizer().lemmatize(word) for word in text]

df['review'] = df['review'].apply(lemmatization)
df['review']
```
